```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Display the data table for preview
```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data_to_load_df = pd.read_csv(city_data_to_load)
ride_data_to_load_df = pd.read_csv(ride_data_to_load)
```

```python
# Display data for city
city_data_to_load_df.head()
```

```python
# Display data for ride
ride_data_to_load_df.head()
```

```python
# Combine the data into a single dataset
# Display the data table for preview
combine_df1 = pd.merge(city_data_to_load_df, ride_data_to_load_df, how='outer', on="city" )

combine_df1.head()
```

```python
# BUBBLE PLOT
```

```python
# ---------------------------------------## ---------------------------------------## ---------------------------------------#
```

```python
# RURAL cities
# Create data frame for RURAL city

rural_df2 = combine_df1.loc[combine_df1.type=="Rural", :]
rural_df2.head()
```

```python
# Crate a data frame with for city and fare

rural_df3 = rural_df2.drop(["driver_count", "date", "ride_id", "type"], axis=1)
rural_df3.head()
```

```python
rural_df4 = rural_df3.groupby(["city"]).mean()
rural_df4
```

```python
rural_df4.max()
```

```python
rural_df4.count()
```

```python
rural_df5 = rural_df2.drop(["fare", "date", "ride_id", "type"], axis=1)
rural_df5.head()
```

```python
rural_df6 = rural_df5.groupby(["city"]).sum()
rural_df6
```

```python
rural_df7 = rural_df2.drop(["driver_count", "fare", "ride_id", "type"], axis=1)
rural_df7.head()
```

```python
rural_df8 = rural_df7.groupby(["city"]).count()
rural_df8
```

```python
plt.scatter(rural_df8, rural_df4, s=rural_df6 * 7, color='yellow', marker="o", label='city types', alpha=1)
plt.show()
```
